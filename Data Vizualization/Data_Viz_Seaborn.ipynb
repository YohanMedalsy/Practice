```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_countries = pd.read_csv('Countries.csv')
```

```python
df_ca_us = df_countries.set_index('Country').loc[['CANADA','UNITED STATES'],'Homicide rate'].reset_index()
df_ca_us
ax = sns.barplot(x="Country", y="Homicide rate", data=df_ca_us).set_title('Homicide rate in Canada vs USA',fontsize=16)
```

It is clear from the above plot that the homicide rate in Canada is much lower than in the US. This confirms current political views.

```python
df_uk_na = df_countries[df_countries['Colonizer'] == 'UK'].set_index('Zone').loc[['North America','Latin America & Caribbean'],['Country','Immigrants (Ratio)','Emigrants (Ratio)']].reset_index()
ax = sns.scatterplot(x = "Emigrants (Ratio)", y = "Immigrants (Ratio)", hue="Country", data = df_uk_na).set_title("Immigrants Ratio vs Emigrants Ratio for Countries colonized by the UK in the continent of America");
plt.legend(bbox_to_anchor=(1., 1), loc=2);
```

From the above plot, one can see that the United States are a major outlier in terms of immigrant ratio vs emigrant ratio. Clearly many people immigrate there and not many leave.
However, my country, Canada, Comes second among all UK colonized countries in the continent of America. It is clear that almost as many people as the US left, but a lot less have immigrated there than to the US.  

```python
df_un = df_countries[['Democracy','Unemployment']]
df_un


dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=dims)
plt.grid(False)
ax = sns.boxplot(x='Democracy', y='Unemployment', data=df_un).set_title('Unemployment vs Democracy',fontsize=16)
```

It is clear form the above plot that increasing the level of democracy in a country decreases the unemployment rate.

```python
sns.jointplot("Urban Population (%)", "HDI", data=df_countries, kind='reg');
```

In the above plot, it is clear that the higher the urban population % the higher the Human development index.

```python
f = plt.figure(figsize=(15,8))
ax = sns.heatmap(df_countries.corr(), annot=True, fmt=".2f").set_title('Correlations among all global variables',fontsize = 20)
```

In the above heatmap, it is clear that the most highly correlated features are the pairs with correlations around 0.7.
