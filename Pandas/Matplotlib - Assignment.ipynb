# Question #1

From the given data, create a trend plot with the following trends on it:
 1. y_3 according to x, color green, no marker, dashed line with width 2, and a legend "I am Legend"
 2. y_2 according to x, color red, no marker, regular line with width 2
 3. y_1 according to x, color skyblue, with blue 'o' markers of size 12. A normal line of width 4.

```python
df = pd.DataFrame({'x': range(1,11), \
                   'y_1': np.random.randn(10), \
                   'y_2': np.random.randn(10) + range(1,11), \
                   'y_3': np.random.randn(10)+range(11,21) })

f = plt.figure()
plt.plot(df['x'], df['y_1'], 'ob', markersize=12)
plt.plot(df['x'], df['y_1'], 'c', linewidth=4.0, label='y_1')
plt.plot(df['x'], df['y_2'], 'r',linewidth=2.0, label='y_2')
plt.plot(df['x'], df['y_3'], '--g',linewidth=2.0, label='I am legend')
plt.legend()
plt.title("Trend Plot of y_1 y_2 y_3")
plt.xlabel('x axis')
plt.show()
```

# Question #3

Print one plot, containing the distribution plots of each of the 3 given distribution. Pass transparency parameters so that all of the parts of each distribution are visible. Add a matching legend.

```python
x1 = np.random.normal(0, 0.6, 1000)
x2 = np.random.normal(-4, 6, 1000)
x3 = np.random.normal(4, 2, 1000)


f = plt.figure()
plt.hist(x1, bins =10,label='x1',alpha=0.8);
plt.hist(x2, bins =10,label='x2',alpha=0.8);
plt.hist(x3, bins =10,label='x3',alpha=0.8);
plt.legend()
plt.title("Histogram Plot of x1 x2 x3")
plt.xlabel('x axis')
plt.show()
```

# Question #4

Create a pie-chart using the given data, to show how many people like each activity. Add a legend and an appropriate title. Locate the legend and the title so that they are not on top of the plot.

Can you read the plot without looking at the data? Can you see if the section "play" is larger than "eat"? What is the ratio between "work" and "study"? These are questions we would like to answer using plots

```python
activities = ['eat', 'sleep', 'work', 'play', 'study'] 
  
slices = [15, 30, 8, 16, 35] 
  
colors = ['r', 'y', 'g', 'b', 'c'] 

f = plt.figure()
plt.pie(slices, colors = colors, labels = activities)
plt.title("Pie chart of daily activities")
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

# Question #5

Correct the plot from Question #4. Add more information to the pie chart so it delivers much more information - a label next to each pie-slice, and the percentage out of all on the pie-slice itself.

```python
f = plt.figure()
plt.pie(slices, colors = colors, labels = activities,autopct='%.2f')
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Pie chart of daily activities")
plt.show()
```

# Question #8

Add a title for the big plot that you created in Question #7. Add a subtitle and axis names for each subplot. Use a tight layout so that the axis names and titles do not overlap.

```python
#subplots
x = np.arange(100)

f, ax = plt.subplots(2, 2,sharey=True, sharex = True)

f.suptitle("Plot of 4 functions",fontsize="x-large")

ax[1,1].plot(x, x**(1/2), color='red')
ax[0,1].plot(x, x/8, color='blue')
ax[1,0].plot(x, x**(1/3), color='green')
ax[0,0].plot(x, x**(1/4), color='yellow')

ax[1,1].title.set_text(' Plot of x^1/2')
ax[0,1].title.set_text('Plot of x/8')
ax[1,0].title.set_text('Plot of x^1/3')
ax[0,0].title.set_text('Plot of x^1/4')

ax[1,1].set_xlabel('x axis')
ax[0,1].set_xlabel('x axis')
ax[1,0].set_xlabel('x axis')
ax[0,0].set_xlabel('x axis')

ax[1,1].set_ylabel('y axis')
ax[0,1].set_ylabel('y axis')
ax[1,0].set_ylabel('y axis')
ax[0,0].set_ylabel('y axis')

f.suptitle("Plot of 4 functions",fontsize="x-large")

plt.tight_layout()
f.subplots_adjust(top=0.85)

plt.show()
```

# Question #9

Read the data from https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv into a pandas dataframe.
 1. Drop all rows with empty values in the columns "RPM", "Type" or "Price"
 2. Plot the relation between the variables "RPM" and "Price" on a scatter plot
 3. Color the data points by the variable "Type"
 4. Add a title and axis names
 5. Add a legend

```python
%matplotlib inline
df_cars = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
df_cars.dropna(subset = ['RPM','Type','Price'], axis = 0,inplace = True)
f = plt.figure()

my_labels = pd.factorize(df_cars['Type'])
plt.scatter(df_cars['RPM'], df_cars['Price'], c = my_labels[0])
plt.colorbar()
    
plt.title("Price vs RPM")
plt.xlabel('RPM')
plt.ylabel('Price')
plt.ylabel('Price')
plt.show()
```
